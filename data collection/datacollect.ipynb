```python
import pandas as pd
import sqlite3 
import datetime
```

```python
player = pd.read_csv("fifa_players.csv")
player
```

```python
#using senven columns and 1000 rows
player = player.loc[0:999,['name','birth_date','height_cm','weight_kgs','positions','nationality','overall_rating']]
```

```python
# Connect to the database
con = sqlite3.connect(r'../database/Football.db')
cursor = con.cursor()
player.to_sql('players', con, index=True, if_exists='replace')
con.commit()
con.close()
```
